### Data SI-erreur

```python
df_si_err = pd.read_excel('data\Climat.xlsx',header=2, nrows=31, skiprows=[3], sheet_name="SI -erreur" )

df_si_err.drop(['Unnamed: 0','Unnamed: 1','Unnamed: 2'], inplace= True, axis= 1)
df_si_err.index = df_si_err.index + 1
df_si_err
```

Pour corriger ce jeu de donnée, nous avons remplacés les données en chaines de caractères présentes avec la moyenne de la température de la veille et du lendemain

```python
for col in df_si_err.columns: 
    for row in df_si_err.index:
        if( type(df_si_err.loc[row,col]) is str):
            df_si_err.loc[row,col] = np.mean([df_si_err.loc[row-1,col],df_si_err.loc[row+1,col]])


df_si_err
```

Les données corrigées et les données sans erreurs sont proches le plus grand écart est de 1,5 degré.
